```python
import re
test_date = '2005-1-1'
# 若要匹配的字符百串为：site sea sue sweet see case sse ssee loses 需要匹配的为以s开头以e 结尾的单词。 正确的正则式为：\bs\S*?e\b
re_result = re.match(r"(.*\b2\d{3}[-/]\d{1,2}[-/]\d{1,2}.*)", test_date)
print(re_result)
```
